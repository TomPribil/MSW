```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
f1 = lambda x: 4*x**3 + 3*x**2 - 2*x + 1
f2 = lambda x: 3*np.sin(2*x-1)
f3 = lambda x: x**3 + np.log(x+1)
```

```python
def bisekce(f, a0, b0, delta=1E-5):
    a, b = a0, b0
    while b**2 - a**2 > 2*delta:
        x = (a + b)/2
        if f(a)*f(x) < 0:
            b = x
        else:
            a = x
    return x


def newton(f, a0, b0, delta=1E-5):
    a, b = a0, b0
    def df(x, h = 1E-3):
        return (f(x+h)-f(x-h))/(2*h)
    xnew = (a+b)/2
    xold = a
    while abs(xnew - xold) > delta:
        xold = xnew
        xnew = xold - f(xold)/df(xold)
    return xnew
```

```python

def plot_result(f, a0, b0, x, title):
    start = time.time()
    plt.title(title)
    plt.plot(x, f(x), "bo")
    end = time.time()
    plt.plot(np.linspace(a0, b0, 100), f(np.linspace(a0, b0, 100)), "r")
    plt.show()
    print("Kořen: ", x)
    print(f"Čas: {str(end-start)} s")
```

```python
a0, b0 = -2, 10
popisek="4*x^3 + 3*x^2 - 2*x + 1"
plot_result(f=f1, a0=a0, b0=b0, x=bisekce(f=f1, a0=a0, b0=b0), title="Bisekce: " +popisek)
plot_result(f=f1, a0=a0, b0=b0, x=newton(f=f1, a0=a0, b0=b0), title="Newton: "+popisek)
```

```python
# Pro funkci 2
popisek = "3*sin(2*x-1)"
a0,b0 = 0.1, 4
plot_result(f=f2, a0=a0, b0=b0, x=bisekce(f=f2, a0=a0, b0=b0), title="Bisekce: " +popisek)
plot_result(f=f2, a0=a0, b0=b0, x=newton(f=f2, a0=a0, b0=b0), title="Newton: "+popisek)
```

```python
# Pro funkci 3
popisek = "x^3 + log(x+1)"
a0, b0 = 0, 15
plot_result(f=f3, a0=a0, b0=b0, x=bisekce(f=f3, a0=a0, b0=b0), title="Bisekce: " +popisek)
plot_result(f=f3, a0=a0, b0=b0, x=newton(f=f3, a0=a0, b0=b0), title="Newton: "+popisek)
```
